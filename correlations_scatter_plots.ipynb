```
AllLDC_production = general_df.groupby('Year')['Production Quantity per capita (kg/cap/year)'].mean()
production_df=pd.DataFrame(AllLDC_production)
production_df.head()
```

```
fa_prod_df = foodavailability_df.merge(production_df, left_index=True, right_index=True)
fa_prod_df.head()
```

```
fa_prod_df.plot(x='Production Quantity per capita (kg/cap/year)', y='Food Availibility per capita (kg/cap/year)', kind='scatter')
plt.xlabel('food production')
plt.ylabel('food availability')
plt.title('relationship between food availability and food production')
```

```
AllLDC_production = general_df.groupby('Year')['Import Quantity (Grains & Ce.)(1000MT)'].mean()
import_df=pd.DataFrame(AllLDC_production)

fa_import_df = foodavailability_df.merge(import_df, left_index=True, right_index=True)
fa_import_df.head()
```

```
fa_import_df.plot(x='Import Quantity (Grains & Ce.)(1000MT)', y='Food Availibility per capita (kg/cap/year)', kind='scatter')
plt.xlabel('food imports')
plt.ylabel('food availability')
plt.title('relationship between food availability and food imports')
```
