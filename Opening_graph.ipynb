```
import pandas as pd
import plotly.express as px
```

```
data_path = '/content/df_final_use.csv' 
frame = pd.read_csv(data_path)
frame
```

```
fig = px.line(frame, x = 'Year', y = 'Food Availibility per capita (kg/cap/year)',
              color = 'Region', line_group = 'Region',
              labels={"Food Availibility per capita (kg/cap/year)": "Food Availibility per capita (kg/cap/year)", "Country": "Country"},
              title = 'Countries: Food Availibility per capita (kg/cap/year)',
              hover_name = 'Country')

#Tidy up the legend
fig.for_each_trace(lambda t: t.update(name=t.name.replace("Country=", "")))
fig.show()
```
